```python
# prompt: script to find similar sentence from list of sentences using keyword weightage

import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from collections import Counter

nltk.download('punkt')
nltk.download('stopwords')

def find_similar_sentences(sentences, query, top_n=3):
  """
  Finds the most similar sentences to a query from a list of sentences based on keyword weightage.

  Args:
    sentences: A list of sentences.
    query: The query sentence.
    top_n: The number of most similar sentences to return.

  Returns:
    A list of the top_n most similar sentences.
  """
  stop_words = set(stopwords.words('english'))

  # Tokenize the query and sentences
  query_tokens = word_tokenize(query)
  query_tokens = [word for word in query_tokens if word.lower() not in stop_words]

  tokenized_sentences = []
  for sentence in sentences:
    tokens = word_tokenize(sentence)
    tokens = [word for word in tokens if word.lower() not in stop_words]
    tokenized_sentences.append(tokens)

  # Calculate keyword weightage
  query_word_counts = Counter(query_tokens)

  similarities = []
  for i, sentence_tokens in enumerate(tokenized_sentences):
    sentence_word_counts = Counter(sentence_tokens)
    common_words = set(query_word_counts.keys()) & set(sentence_word_counts.keys())
    similarity = sum([min(query_word_counts[word], sentence_word_counts[word]) for word in common_words])
    similarities.append((i, similarity))

  # Sort sentences by similarity
  similarities.sort(key=lambda x: x[1], reverse=True)

  # Return top_n most similar sentences
  top_sentences = [sentences[i] for i, _ in similarities[:top_n]]
  return top_sentences

# Example usage
sentences = [
    "This is a sample sentence.",
    "Another sentence with different words.",
    "This sentence is similar to the first one.",
    "A completely unrelated sentence."
]
query = "This sentence is similar to the first one."

similar_sentences = find_similar_sentences(sentences, query)
print(similar_sentences)
```
